```python
import os
import random
import math

following_iter = False
W = []  # weight matrix
Y = []
max_iterations = 350
nums = []
used_neurons = 0
actual = 0



max_side_length = 512
side_of_num = 7
coding_symbols = {'@': 1, '-': -1}
```

```python
def load_pic():
    num_list = []
    for fname in os.listdir("D:/mrz2\example"):
        fpath = os.path.join("example", fname)
        if fname.endswith(".txt"):
            with open(fpath) as fp:
                file_content = fp.read(max_side_length)
                file_content = str.replace(file_content, "\n", "")
                file_content = str.replace(file_content, "\r", "")
                number = []
                for c in file_content:
                    number.append(coding_symbols[c])
                num_list.append(number)
    return num_list


def char_for(x):
    symbols_for_decoding = {1: '@', -1: '-'}
    return symbols_for_decoding[x]


def print_num(image, size):
    i = 0
    for symb in image:
        print(f"{char_for(symb)}", end='')
        i += 1
        if i % size == 0:
            print('\n')


def working_with_neurons(side):
    global used_neurons, W
    neurons = int(math.pow(side, 2))
    used_neurons = neurons
    for i in range(0, used_neurons):
        W.append([0 for _ in range(0, used_neurons)])


def learning(number):
    global used_neurons, W, nums
    nums.append(number)
    X = number
    for i in range(0, used_neurons):
        for j in range(0, used_neurons):
            if i != j:
                W[i][j] = W[i][j] + X[i] * X[j]
            else:
                W[i][j] = 0


def recovering(number, side):
    global Y, max_iterations
    Y = number
    iter = 0
    print('Процесс восстановления по нейронам: ')
    while nums.count(Y) == 0:
        steps_of_recovering(side)
        iter = iter + 1
        if actual >= max_iterations:
            print(actual)
            return False, Y, iter
    return True, Y, iter


def steps_of_recovering(side):
    global actual, Y, used_neurons, following_iter
    sign = lambda x: -1 if x < 0 else 1  # задаем функцию активаци
    diapason = random.randrange(0, used_neurons, 1)
    net = 0
    for i in range(0, used_neurons):
        net = net + Y[i] * W[i][diapason]
    signnet = sign(net)
    if signnet == Y[diapason]:
        actual += 1
    else:
        print(f'Нейрон {diapason}: с {Y[diapason]}  на  {signnet}')
        Y[diapason] = signnet
        if following_iter:
            print_num(Y, side)
            print('\n')
        actual = 0
```

```python
def main():
    numbers = load_pic()
    with open("modified.txt") as f:
        contents = f.read(max_side_length)
        contents = str.replace(contents, "\n", "")
        contents = str.replace(contents, "\r", "")
        numb = []
        for sign in contents:
             numb.append(coding_symbols[sign])

    print("Картинки для обучения:")
    for i in numbers:
        print_num(i, side_of_num)
        print()

    print("Процесс обучения")
    working_with_neurons(side_of_num)
    for each_list in numbers:
        learning(each_list)
    print()

    print("Поломанная картинка")
    print_num(numb, side_of_num)
    print()

    print("Процесс восстановления")
    (recognize, wrong_num, iters) = recovering(numb, side_of_num)
    print()

    if recognize:
        print(f"Буква была восстановлена за {iters} итераций:")
    else:
        print(f"Ошибка! Номер проблемной итерации: {iters} ")
    print_num(wrong_num, side_of_num)


if __name__ == '__main__':
    main()
```
